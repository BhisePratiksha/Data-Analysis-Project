*Outcome of Incident wise count*

```python
sales_gen = df.groupby(['Outcome of Incident'], as_index=False)['Count'].sum().sort_values(by='Count', ascending=False)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x = 'Count',y= 'Outcome of Incident' ,data = sales_gen)
```

```python
sales_gen = df.groupby(['Million Plus Cities'], as_index=False)['Count'].sum().sort_values(by='Count', ascending=False)

sns.set(rc={'figure.figsize':(15,30)})
sns.barplot(x = 'Count',y= 'Million Plus Cities' ,data = sales_gen)
```

The data clearly indicates that Chennai reports the highest number of accident cases compared to other cities in the dataset.

```python
sales_gen = df.groupby(['Cause category'], as_index=False)['Count'].sum().sort_values(by='Count', ascending=False)

ax = sns.countplot(x = 'Cause category',data = df)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x = 'Cause category',y= 'Count' ,data = sales_gen)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Cause Subcategory'], as_index=False)['Count'].sum().sort_values(by='Count', ascending=False)

ax = sns.countplot(x = 'Cause Subcategory',data = df)
    
sns.set(rc={'figure.figsize':(70,5)})
sns.barplot(x = 'Cause Subcategory',y= 'Count' ,data = sales_gen)
for bars in ax.containers:
    ax.bar_label(bars)
```

The data clearly indicates that Chennai reports the highest number of accident cases compared to other cities in the dataset.
